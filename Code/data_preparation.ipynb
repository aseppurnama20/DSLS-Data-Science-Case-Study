```python
def q25(x):
    return x.quantile(0.5)


def q50(x):
    return x.quantile(0.5)

def q75(x):
    return x.quantile(0.75)
```

```python
alerts = pd.read_csv('gdrive/My Drive/DS DSLS/aggregate_alerts_Kota Cirebon.csv')
irregularities = pd.read_csv('gdrive/My Drive/DS DSLS/aggregate_median_irregularities_Kota Cirebon.csv')
jams = pd.read_csv('gdrive/My Drive/DS DSLS/aggregate_median_jams_Kota Cirebon.csv')
```

```python
dg1 = alerts.groupby('street')
```

```python
## group metrics
fun1 = ['count', np.mean, np.std, q25, q50, q75]
fun1_dict = {}
col_f1 = ['street']
for cols in ['total_records']:
    fun1_dict[cols] = fun1
    for met in ['cnt','avg','std', 'p25', 'p50','p75']:
        col_f1.append(cols+'_'+'alerts'+'_'+met)
        
loc_f1 = dg1.agg(fun1_dict).reset_index()
loc_f1.columns = col_f1
```

```python
loc_f1.sample(3)
```

```python
irregularities.info()
```

```python
dg2 = irregularities.groupby('street')
```

```python
## group metrics
fun2_1 = ['count', np.mean, np.std, q25, q50, q75]
fun2_1_dict = {}
col_f2_1 = ['street']
for cols in  ['median_length','median_delay_seconds', 'median_regular_speed', 'total_records', 'median_seconds', 'median_speed', 'median_jam_level'] :
    fun2_1_dict[cols] = fun2_1
    for met in ['cnt','avg','std', 'p25', 'p50','p75']:
        col_f2_1.append(cols+'_'+'irr'+'_'+met)
        
loc_f2_1 = dg2.agg(fun2_1_dict).reset_index()
loc_f2_1.columns = col_f2_1
```

```python
loc_f2_1.sample(3)
```

```python
jams.info()
```

```python
dg3 = jams.groupby('street')
```

```python
## group metrics
fun3_1 = ['count', np.mean, np.std, q25, q50, q75]
fun3_1_dict = {}
col_f3_1 = ['street']
for cols in  ['median_length','median_delay', 'median_speed_kmh', 'total_records', 'median_level'] :
    fun3_1_dict[cols] = fun3_1
    for met in ['cnt','avg','std', 'p25', 'p50','p75']:
        col_f3_1.append(cols+'_'+'jams'+'_'+met)
        
loc_f3_1 = dg3.agg(fun3_1_dict).reset_index()
loc_f3_1.columns = col_f3_1
```

```python
loc_f3_1.sample(3)
```

```python
jams_target = jams[['date', 'time', 'street', 'geometry', 'level']]
```

Merge Feature

```python
feature = [loc_f1, loc_f2_1, loc_f3_1]
```

```python
df_all = jams_target.copy()
for d in feature:
    df_all = pd.merge(df_all, d, on = 'street', how = 'left')
```
