# Handling Missing Value

```python
col_null = data.isnull().sum()/len(data)
col_null[col_null>0]
```

```python
data2 = data.fillna(0)
```

```python
data2.sample(10)
```

# Handling Outlier

```python
col_out = set(data2.columns)-set(['date', 'time', 'street', 'geometry','level'])
```

```python
outlier_rules = data2[col_out].quantile(0.98).to_dict()
```

```python
data3 = data2.copy()
for c in col_out:
    data3[c] = np.where(data3[c]>=outlier_rules[c], outlier_rules[c], data3[c])
```
